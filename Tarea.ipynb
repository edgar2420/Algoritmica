```python
from sklearn import neighbors
from sklearn.model_selection import train_test_split


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('AppsandOffers_2022_SchoolLevel.csv')
```

```python
df.drop('school_name', axis=1, inplace=True)
df.drop('administration', axis=1, inplace=True)
df.drop('city', axis=1, inplace=True)
df.drop('branch', axis=1, inplace=True)
df.drop('status', axis=1, inplace=True)
df.drop('_merge', axis=1, inplace=True)
df.drop('second_foreign_lang', axis=1, inplace=True)

```

```python
numeric_columns = df.select_dtypes(include=np.number)
corr = df.corr()
corr.style.background_gradient()
```

```python
# Inicializar el modelo
k_neighbors = neighbors.KNeighborsClassifier()
```

```python
# Preparar la data para entrenar el modelo
data_x = df.drop(['id', 'num_subscribers', 'avg_rating', 'num_reviews', 'num_comments','num_lectures','content_length_min'], axis=1)
data_y = df.num_reviews
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.1)
```

```python
# Entrenar el modelo
k_neighbors.fit(train_x, train_y)
```

```python
# Medir la confianza de nuestro modelo
print(f'Score: {round(k_neighbors.score(test_x, test_y), 4) * 100}%')
```
